### Reformating genres,keywords,cast,crew

```python
# convert genres to ["Action"Comedy"...]
```

```python
import ast #for converting string to list
```

```python
def convert(obj): #function that maps each genre and extracts name fron each object...
    A=[]
    for x in ast.literal_eval(obj):
        A.append(x['name'])
    return A
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert) #similarly filtering keywords
```

```python
movies['keywords']
```

```python
movies['cast']
```

```python
def convert3(obj):#we will be taking only first 3 names from cast array
    A = []
    counter = 0
    for x in ast.literal_eval(obj):
        if counter != 3:
            A.append(x['name'])
            counter+=1
        else:
            break
    return A
```

```python
movies['cast']= movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def findDir(obj):#extracting names from crew where job is director
    A=[]
    for x in ast.literal_eval(obj):
        if x['job'] =='Director':
            A.append(x['name'])
            break
    return A
```

```python
movies['crew']=movies['crew'].apply(findDir)
```

```python
movies.head()
```

```python
#converting overview from string format to list so that we can concatenate with others
movies['overview']=movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

#### Removing spaces

```python
#removing spaces between words Sam Worthington-->SamWorthington
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

### forming tags

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
#converting list to string so that we can do s=text vectorization.
new_dataset['tags']=new_dataset['tags'].apply(lambda x:" ".join(x))
```

STEMMING OF TAGS

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron')
```

```python
new_dataset['tags']=new_dataset['tags'].apply(stem)
```

CREATING VECTORS

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_dataset['tags']).toarray()
```

```python
vectors[0]
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import euclidean_distances
distances=euclidean_distances(vectors,vectors)
```

```python
distances[0]
```

```python
sorted(list(enumerate(distances[0])),key=lambda x:x[1])
```

```python
def recommend(movie):
    movie_index=new_dataset[new_dataset['title']== movie].index[0]
    node_distances=distances[movie_index]
    recommended_list=sorted(list(enumerate(node_distances)),key=lambda x:x[1])[0:6]
    temp=[];
    for i in recommended_list:
            temp.append(new_dataset.iloc[i[0]].title)
    return temp
```
